1. Пожалуйста, реализуйте функцию ```sum_error(y_true, y_pred)```, которая будет вычислять сумму отклонений:

$$Error = \sum_{i}^n (y_i^{true}-y_i^{pred})$$

````python
import numpy as np

def sum_error(y_true, y_pred):
    #Начало вашего кода
    return sum(y_true)-sum(y_pred)

    #Конец вашего кода

np.random.seed(1)
y_true = np.random.randint(0,10,(10,))
y_pred = np.random.randint(0,10,(10,))
print(y_true)
print(y_pred)
print(sum_error(y_true, y_pred))
````

2. Пожалуйста, реализуйте функцию ```sum_square_error(y_true, y_pred)```, которая будет вычислять сумму  квадратов отклонений:

$$Error = \sum_{i}^n (y_i^{true}-y_i^{pred})^2$$

````python
def sum_square_error(y_true, y_pred):
    #Начало вашего кода
    return sum((y_true-y_pred)**2)

    #Конец вашего кода

print(y_true)
print(y_pred)
print(sum_square_error(y_true, y_pred))
````

3. Пожалуйста, реализуйте функцию ```mean_square_error(y_true, y_pred)```, которая будет вычислять среднеквадратические отклонение:

$$Error = \frac {1}{n} \sum_{i}^n (y_i^{true}-y_i^{pred})^2$$

````python
def mean_square_error(y_true, y_pred):
    #Начало вашего кода
    return ((y_true-y_pred)**2).mean()
    #Конец вашего кода

print(y_true)
print(y_pred)
print(mean_square_error(y_true, y_pred))
````

5. Пожалуйста, реализуйте функцию ```split_data(data, ratio=0.1)```, которая разбивает набор данных, загруженный в предыдущем вопросе, на наборы данных ```train``` и ```validation``` в соотношении 9:1.

````python
len(num_data)
````

````python
def split_data(data, ratio=0.1):
    np.random.seed(1)
    #Начало вашего кода    
    indicies=np.arange(0,len(data))
    np.random.shuffle(indicies)
    val_idx=indicies[:round(ratio*len(indicies))]
    train_idx=indicies[round(ratio*len(indicies)):]
    val_data=data.loc[val_idx]
    train_data=data.loc[train_idx]

    #Конец вашего кода
    return {"validation":val_data, "train":train_data}

splitted_data = split_data(num_data, ratio=0.1)
````

8. Теперь пришло время использовать модель линейной регрессии для каждого параметра в списке ```important``` для прогнозирования ```SalePrice```. Пожалуйста, реализуйте функцию ```train_simple_lin_regressors(data, important, target="SalePrice")```, которая будет обучать модель линейной регрессии для каждого параметра в списке ```important```. Функция должна возвращать словарь линейных регрессоров с параметрами, используемыми для обучения в качестве ключей.

````python
#Начало вашего кода 
from sklearn.linear_model import LinearRegression
def train_simple_lin_regressors(data, important, target="SalePrice"):
    y=data[target].values
    dic={}
    for col in important:
        X=data[col].values
        dic[col]=LinearRegression().fit(X.reshape(-1, 1),y)
    return dic
#Конец вашего кода

simple_regressors = train_simple_lin_regressors(splitted_data["train"], important, target="SalePrice")
print(simple_regressors)
````

### Ожидаемые результаты:

9. Спрогнозируйте ```"SalePrice"``` для набора данных ```validation``` с использованием обученных моделей линейной регрессии. 
Пожалуйста, реализуйте функцию ```validation(regressors, data, target="SalePrice")```, которая будет принимать словарь моделей линейной регрессии ```regressors```, данные ```validation``` и параметр ```target```, который должен быть предсказан. Функция должна возвращать ```DataFrame``` с первым столбцом, который будет истинным ```"SalePrice"```,  а остальные столбцы предсказаны моделями линейной регрессии.

````python
def validation(regressors, data, target="SalePrice"):
    #Начало вашего кода
    data1=pd.DataFrame(splitted_data["validation"]['SalePrice'])
    
    for i in simple_regressors:
        data1[i]=simple_regressors[i].predict(data[i].values.reshape(-1, 1))
    return data1
    #Конец вашего кода

pd_predictions = validation(simple_regressors, splitted_data["validation"], target="SalePrice")

pd_predictions
````
