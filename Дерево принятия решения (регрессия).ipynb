2. Пожалуйста, реализуйте функцию ```split_data(data, ratio=0.1)```, которая разбивает набор данных, загруженный в предыдущем вопросе, на наборы данных ```train``` и ```validation``` в соотношении 9:1.

````python
import numpy as np

def split_data(data, ratio=0.1):
    np.random.seed(1)
    #Начало вашего кода    
    val = ratio
    train = 1 - val
    
    indices = np.arange(0, len(data))
    np.random.shuffle(indices)
    
    val_indices = indices[:round(val * len(data))]
    train_indices = indices[round(val * len(data)):]
    
    val_data = data.iloc[val_indices]
    train_data = data.iloc[train_indices]
    #Конец вашего кода
    return {"validation":val_data, "train":train_data}

splitted_data = split_data(num_data, ratio=0.1)
````

Функция должна возвращать словарь с обученным деревом принятия решения (регрессии), списком ```parameters```.

````python
#Начало вашего кода 
from sklearn.tree import DecisionTreeRegressor
import numpy as np

def train_decision_tree_regressor(data, parameters, target = "SalePrice", degree = 1, random_state=0):
    regressors = {}
    regressor = DecisionTreeRegressor(random_state = random_state)
    
    #X = np.reshape(data[parameters].values,(-1,1))
    y = data[target].values
    regressor.fit(data[parameters[1:]].values,y)
    
    regressors["regressor"] = regressor
    regressors["parameters"] = parameters[1:]

    return regressors
    
#Конец вашего кода
model = train_decision_tree_regressor(splitted_data["train"], important, target="SalePrice")
print(model)
````

Пожалуйста, реализуйте функцию ```validation(model, data, target="SalePrice")```, которая будет принимать словарь ```model```, данные ```validation``` и параметр ```target```, который должен быть предсказан. Функция должна возвращать ```DataFrame``` с первым столбцом, который будет истинным ```"SalePrice"```,  а второй столбец предсказан деревом принятия решения и ```rmse```.

````python
#Начало вашего кода
def validation(model,data,target = "SalePrice"):
    pd_val = pd.DataFrame(data, columns = [target])
    col_pred = ""
    for i in model["parameters"]:
        col_pred = col_pred + f"{i}-"
    col_pred = col_pred[:-1]
    dec_tree_features = data[model["parameters"]].values
    pd_val[col_pred] = model["regressor"].predict(dec_tree_features).astype(int)
    rmse = (np.mean((data[target]-pd_val[col_pred])**2)**(1/2))
    return pd_val,rmse
#Конец вашего кода
````

**Примечание:** figsize=(8,8)

````python
#Начало вашего кода
fig = plt.figure(figsize = (8,8))

pd_predictions_train, rmse_train = validation(model, splitted_data["train"], target="SalePrice")

sns.set(style = "white", context = "notebook")

plt.scatter(pd_predictions_train.iloc[:,0],pd_predictions_train.iloc[:,1],
           color = "tab:blue")
plt.title(f"Train dataset RMSE {rmse_train:.2f} USD")
plt.xlabel("True Sale Price")
plt.ylabel("Predicted Sale Price");

#Конец вашего кода
````

9. Что вы думаете о такой большой разнице в ```rmse``` для набора данных ```train``` и ```validation```?

Модель Decision Tree обучалась на train dataset, поэтому вполне закономерно маленькое значение rmse и высокая точность модели. Но в validation dataset, есть значения, которые не учтены в train dataset, поэтому и ошибка довольно большая.  

Думаю, что одна из причин большой ошибки в том, что dataset разбивается просто перемешкой, тогда как, мне кажется, было бы правильнее разбивать весь dataset на страты и брать по нескольку представителей от каждой страты в validation dataset.  

Также, нужно учитывать саму специфику Decision Tree. В отличии от линейной регрессии, которая старается описать весь dataset функцией, Decision Tree просто сравнивает новые исходые данные со старыми. И она не учитывает outliers. Таким образом, приведу пример:  

Age Salary  
18   10000  
19   12000  
20   25000  
22   18000  
25   24000  
30   30000  

Модель линейной регрессии попытается описать данные функцией Salary от Age и график этой функции будет стремиться к прямой линии, какой бы ни была степень полинома, потому что в зависимости от возраста, зарплата растет линейно, за исключением 20 лет, где зарплата аномальна. Функция плавно постарается охватить эту точку, но ей это не удастся. После того, как мы получим функцию вида $ax+b$ или $ax^2+bx+c$, при подстановке 20 вместо x, мы не получим y в 25000.
Линейная функция будет $y = 1444,1x - 12418$. Подставив 20, мы получим y лишь 16464.  

В случае с полиномом третьей степени, $y = 49,48x^3-3638x^2+89021x-702277$, подставив 20 мы получим 18783.  

В случае же с DecisionTree, модель обучится на таких данных с аномальными значениями:	
............................>21			
................>19..........................>25	
...........>18........25000......>24.......30000  
10000.......12000......18000....24000	

Когда мы подставляем значения больше 19, но меньше 21 в validation dataset, наша модель всегда будет выдавать аномальное значение 25000, на котором она обучилась, тогда как в оригинальном validation dataset значение будет более реалистичным, скажем 15000, и тогда наша модель Decision Tree будет выдавать большую ошибку, как в нашем примере в задании 8. Получается, что Decision Tree гораздо в большей степени восприимчив к аномальным значениям, чем линейная регрессия, поэтому эта модель должна быть использована лишь в случаях полной уверенности в правильности данных.
