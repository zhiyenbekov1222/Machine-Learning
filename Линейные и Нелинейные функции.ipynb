2. –í–∞–º –¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è y, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –æ–Ω–∞ –ª–∏–Ω–µ–π–Ω–æ–π ```linear``` –∏–ª–∏ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–π ```non-linear```. –ï—Å–ª–∏ –≤–∞–º –Ω—É–∂–Ω–æ, –≤—ã –º–æ–∂–µ—Ç–µ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —ç—Ç–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –ø–µ—Ä–µ–∫—Ä–µ—Å—Ç–Ω–æ–π –ø—Ä–æ–≤–µ—Ä–∫–∏.

    * $y=x+1/x-20$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y1```.
    
    * $y=x^2/x-5x$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y2```.
    
    * $y=1/log(x)$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y3```.
    
    * $y=sin(x)+0.3x^2-x^4$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y4```.

    * $y=(x-4)^2$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y5```.
    
    * $y=(x^2-4)/(x-2)$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y6```.
    
    * $y=(x^3-8)/(x^2-2)$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y7```.
    
    * $y=1-x$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y8```.
    
    * $y=0$ –ü—Ä–∏—Å–≤–æ–π—Ç–µ —Å–≤–æ–π –æ—Ç–≤–µ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```y9```.


–ü—Ä–∏–º–µ—Ä –æ—Ç–≤–µ—Ç–∞: ```y="linear"```  (string)

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
y1 = "non-linear"
y2 = "linear"
y3 = "non-linear"
y4 = "non-linear"
y5 = "non-linear"
y6 = "linear" # –ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º —á—Ç–æ (x^2 - 4) -> (x - 2)(x + 2) –∑–∞—Ç–µ–º  (x - 2)(x + 2) / (x - 2) = x + 2
y7 = "linear" # –ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º —á—Ç–æ –ø–æ —Ñ–æ—Ä–º—É–ª–µ (x^3 - 8) -> (x - 2)(x^2 + 2x + 4) / (x - 2)(x + 2) –¥–∞–ª–µ–µ  (x - 2) —Å–æ–∫—Ä–∞—â–∞—é—Ç—Å—è –æ—Å—Ç–∞–µ—Ç—Å—è
# (x^2 + 2x + 4) / (x + 2) -> –ø–æ —Ñ–æ—Ä–º—É–ª–µ (x + 2)^2 / (x + 2) –ø—Ä–∏ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–∏ —Å—Ç–µ–ø–µ–Ω–∏ –≤—ã—á–∏—Ç–∞—é—Ç—Å—è –æ—Å—Ç–∞–µ—Ç—Å—è x + 2
y8 = "linear"
y9 = "linear"
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
````

3. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é ```rate_of_change(x1,x2,y1,y2)```, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —Å–∫–æ—Ä–æ—Å—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è (rate of change = RoC) –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å 4 –ø–∞—Ä–∞–º–µ—Ç—Ä–∞, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö 2 x-–∑–Ω–∞—á–µ–Ω–∏—è –∏ 2 y-–∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è 2 —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–µ–∫: ```x1, x2, y1, y2```:

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
def rate_of_change(x1,x2,y1,y2):
    return (y2 - y1) / (x2 - x1)
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

print(f"RoC: {rate_of_change(0,1,0,1)}")
print(f"RoC: {rate_of_change(0,1,1,0)}")
print(f"RoC: {rate_of_change(10,3,-10,4)}")
````

4. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é ```plot_secant(fun,x1,x2,xlims=(-10,10),ylims=(-100,200))```, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é ```fun``` –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ xlims –ø–æ –æ—Å–∏ x –∏ ylims –ø–æ –æ—Å–∏ y. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, —Å–ª–µ–¥—É–µ—Ç –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å–µ–∫—É—â—É—é –ª–∏–Ω–∏—é (secant), —Å–æ–µ–¥–∏–Ω—è—é—â—É—é —Ç–æ—á–∫–∏ ```x1``` –∏ ```x2``` (–∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏, –∫–∞–∫ –≤ –≤–∏–¥–µ–æ–ª–µ–∫—Ü–∏–∏).

    **–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** –≤–∞–º –º–æ–∂–µ—Ç –ø–æ–Ω–∞–¥–æ–±–∏—Ç—å—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é ```rate_of_change()```.

````python
#–ù–∞—á–∞–ª–æ –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def plot_secant(fun,x1,x2,xlims=(-10,10),ylims=(-100,200)):
    y1 = fun(x1)
    y2 = fun(x2)
    m = rate_of_change(x1, x2, y1, y2)
    b = y1 - m * x1
    
    lsx = np.linspace(xlims[0],xlims[1], 100)
    lsy = []
    for num in lsx.tolist():
        lsy.append(fun(num))  
    
    sec = []
    for non in lsx.tolist():
        sec.append(m * non + b)
    plt.grid();
    plt.plot(lsx, np.array(lsy), label="function");
    plt.plot(lsx, np.array(sec), label="y = " + str(round(m, 1)) +"x + " + str(round(b, 1)));
    plt.title("x1 = " + str(x1) + " & x2 = " + str(x2));
    plt.legend();
    plt.xlim(xlims);
    plt.ylim(ylims);
#–ö–æ–Ω–µ—Ü –≤–∞—à–µ–≥–æ –∫–æ–¥–∞

y = lambda x: -x**2+2*x+20
plot_secant(y,-5,3)
````

### –û–∂–∏–¥–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:




````python
t = lambda x: -x**2+10
x2 = [5, 3, 1, 0, -1, -2, -3, -4, -4.9]
plt.figure(figsize=(15, 15));
for idx in range(len(x2)):    
    plt.subplot(3,3, idx + 1);
    plot_secant(t,-5,x2[idx], xlims=(-10,10),ylims=(-100,200));  
        
````
