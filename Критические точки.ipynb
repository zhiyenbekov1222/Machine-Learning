1. Пожалуйста, найдите экстремальные точки данной функции $y$. Создайте два пустых списка ```relmin``` и ```relmax```. Если критическая точка является относительным минимумом, добавьте ее в список ```relmin```. Если критическая точка является относительным максимумом, добавьте ее в список ```relmax```.

    * $y=-x^3+2x^2$
    
**Примечание:** уклон в критических точках равен нулю.

````python
#Начало вашего кода
relmin = []
relmax = []

# Находим производную 
# y = -3x^2 + 4x
# Находим дискреминант
# x1 = -4 + sqrt(16 - 4 * (-3) * 0) / 2 * (-3)
# x1 = (-4 + 4) / -6 = 0
# x2 = -4 - sqrt(16 - 4 * (-3) * 0) / 2 * (-3)
# x2 = (-4 - 4) / 2 * (-3) = -8 / -6 = 4 / 3
# Находим максимум и минимум
# функции возрастает на интервале (0; 4/3), в точке x = 0 функция достигает минимума а в точке x = 4/3 достигает максимума
# Я провел экспиремент взял x = -1, x = 1, x = 2 и подставил в производную, при x = -1 и x = 2, y < 0 (функция убывает), а при 
# значении x = 1 y > 0 (функция возрастает входит в интервал (0; 4/3))
relmin.append(0)
relmax.append(4/3)
#Конец вашего кода
print(relmin)
print(relmax)
````

5. Пожалуйста, реализуйте функцию ```plot_f_df(f,df,xlims=(-10,10))```, которая будет отображать функцию ```f``` и производную ```df``` в диапазоне xlims по оси x.

````python
#Начало вашего кода
import matplotlib.pyplot as plt
%matplotlib inline


def plot_f_df(f,df,xlims=(-10,10)):
    lsx = np.linspace(xlims[0], xlims[1])
    lsy = list(map(f, lsx.tolist()))
        
    lsy_2 = list(map(df, lsx.tolist()))
    plt.figure(figsize=(10, 5));
    # f plots    
    plt.subplot(1,2,1);
    plt.grid();
    plt.plot(lsx, lsy, label="f");
    plt.legend(fontsize="16");
    plt.xlim(xlims);
    
    # df plots
    plt.subplot(1,2,2);
    plt.grid();
    plt.plot(lsx, lsy_2, label="df", linestyle="-.", color="r");
    plt.legend(fontsize="16");
    plt.xlim(xlims);
#Конец вашего кода

plot_f_df(lambda x: -x**2+10*x, lambda x: -2*x+10)
````
